```python
# Representasi graf tak berarah menggunakan adjacency list (revisi)
graf = {
    'A': ['T', 'P'],
    'B': ['U', 'R', 'W'],
    'C': ['T', 'I', 'F'],
    'D': ['S', 'A'],
    'E': ['B', 'I']
}
print_graph(graf)

# Fungsi untuk menampilkan graf (jawaban no 1)
def print_graph(graf):
    print("Representasi graf:")
    for simpul in graf:
        print(f"{simpul} ~-~-~-> {graf[simpul]}")
    print()

bfs(graf, 'A')

# Fungsi BFS (Breadth-First Search) (jawaban no 2)
from collections import deque
def bfs(graf, start):
    print(f"Hasil traversal BFS dari simpul {start}:")
    visited = set()
    queue = deque([start])

    while queue:
        simpul = queue.popleft()
        if simpul not in visited:
            print(simpul, end=" ")
            visited.add(simpul)
            queue.extend([tetangga for tetangga in graf.get(simpul, []) if tetangga not in visited])
    print("\n")

print("Hasil traversal DFS dari simpul A:")
dfs(graf, 'A')
print("\n")

# Fungsi DFS (Depth-First Search) (jawaban no 3)
def dfs(graf, start, visited=None):
    if visited is None:
        visited = set()
    if start not in visited:
        print(start, end=" ")
        visited.add(start)
        for tetangga in graf.get(start, []):
            dfs(graf, tetangga, visited)

print("Hasil traversal DFS dari simpul A:")
dfs(graf, 'A')
print("\n")

# Modifikasi BFS yang mengembalikan list (jawaban no 4)
def bfs_list(graf, start):
    visited = set()
    queue = deque([start])
    hasil = []

    while queue:
        simpul = queue.popleft()
        if simpul not in visited:
            hasil.append(simpul)
            visited.add(simpul)
            queue.extend([tetangga for tetangga in graf.get(simpul, []) if tetangga not in visited])

    return hasil
print("Urutan kunjungan BFS (sebagai list) dari A:")
print(bfs_list(graf, 'A'))
print()

print("Jalur dari E ke R (menggunakan DFS):")
print(find_path(graf, 'E', 'R'))
print()

def find_path(graf, start, end, path=None):
    if path is None:
        path = []
    path = path + [start]
    if start == end:
        return path
    if start not in graf:
        return None
    for tetangga in graf[start]:
        if tetangga not in path:
            jalur = find_path(graf, tetangga, end, path)
            if jalur:
                return jalur
    return None

# Fungsi untuk mengecek apakah graf terhubung (jawaban no 5)
def is_connected(graf):
    simpul_awal = next(iter(graf))
    visited = set()

    def dfs_kunjungi(simpul):
        if simpul not in visited:
            visited.add(simpul)
            for tetangga in graf.get(simpul, []):
                dfs_kunjungi(tetangga)

    dfs_kunjungi(simpul_awal)
    semua_simpul = set(graf.keys())
    # Tambahkan semua simpul yang ada di daftar tetangga juga
    for daftar in graf.values():
        semua_simpul.update(daftar)

    return visited == semua_simpul

print("Apakah graf terhubung sepenuhnya?")
print(is_connected(graf))
```
