```python
class AntrianBioskop:
    def __init__(self, kapasitas_maks=10):
        self.antrian = []
        self.kapasitas_maks = kapasitas_maks

    def tambah_antrian(self, kode_pelanggan):
        if "7" in kode_pelanggan:
            print(f"Pelanggan VIP {kode_pelanggan} langsung mendapatkan kursi tanpa antri.")
            return
        if len(self.antrian) >= self.kapasitas_maks:
            print(f"Antrian penuh! Pelanggan {kode_pelanggan} harus menunggu.")
            return
        self.antrian.append(kode_pelanggan)
        print(f"Pelanggan {kode_pelanggan} masuk antrian.")

    def layani_pelanggan(self):
        if self.kosong():
            print("Antrian kosong! Tidak ada pelanggan untuk dilayani.")
            return None
        pelanggan_dilayani = self.antrian.pop(0)
        print(f"Pelanggan {pelanggan_dilayani} telah dilayani.")
        return pelanggan_dilayani

    def pelanggan_depan(self):
        return self.antrian[0] if not self.kosong() else "Antrian kosong"

    def pelanggan_belakang(self):
        return self.antrian[-1] if not self.kosong() else "Antrian kosong"

    def kosong(self):
        return len(self.antrian) == 0

    def ukuran_antrian(self):
        return len(self.antrian)

    def __str__(self):
        return "Antrian Saat Ini: " + " <- ".join(self.antrian) if not self.kosong() else "Antrian kosong"

antrian = AntrianBioskop()

while True:
    print("<<<<<<<<<<<<<>>>>>>>>>>>>>>>")
    print("SISTEM ANTRIAN TIKET BIOSKOP")
    print("<<<<<<<<<<<<<>>>>>>>>>>>>>>>")
    print("1. Tambah pelanggan ke dalam antrian")
    print("2. Layani pelanggan berikutnya (hapus dari antrian)")
    print("3. Tampilkan daftar pelanggan dalam antrian")
    print("4. Lihat pelanggan pertama dan terakhir dalam antrian")
    print("5. Periksa jumlah pelanggan dalam antrian")
    print("6. Keluar dari sistem")


    pilihan = input("Pilih menu (1-6): ")

    if pilihan == "1":
        kode_pelanggan = input("Masukkan kode pelanggan (contoh: ZU57): ").upper()
        antrian.tambah_antrian(kode_pelanggan)

    elif pilihan == "2":
        antrian.layani_pelanggan()

    elif pilihan == "3":
        print(antrian)

    elif pilihan == "4":
        print(f"Pelanggan di depan: {antrian.pelanggan_depan()}")
        print(f"Pelanggan di belakang: {antrian.pelanggan_belakang()}")

    elif pilihan == "5":
        print(f"Ukuran antrian saat ini: {antrian.ukuran_antrian()}")

    elif pilihan == "6":
        print("Keluar dari sistem. Terima kasih.")
        break

    else:
        print("Pilihan tidak valid! Silakan masukkan angka 1-6.")
```

```python
class AntrianFotokopi:
    def __init__(self):
        self.antrian = []

    def tambah_pesanan(self, id_pesanan, jumlah_halaman):
        if jumlah_halaman < 10:
            print(f"Pesanan {id_pesanan} Fast Track! Langsung diproses tanpa antri.")
            return
        elif jumlah_halaman > 100:
            self.antrian.append((id_pesanan, jumlah_halaman // 2))  # Tahap 1
            self.antrian.append((id_pesanan, jumlah_halaman - jumlah_halaman // 2))  # Tahap 2
            print(f"Pesanan {id_pesanan} ({jumlah_halaman} halaman) diproses dalam dua tahap.")
        else:
            self.antrian.append((id_pesanan, jumlah_halaman))
            print(f"Pesanan {id_pesanan} ({jumlah_halaman} halaman) masuk antrian.")

    def layani_pesanan(self):
        if self.apakah_kosong():
            print("Antrian kosong! Tidak ada pesanan untuk diproses.")
            return None
        pesanan_dilayani = self.antrian.pop(0)
        print(f"Pesanan {pesanan_dilayani[0]} ({pesanan_dilayani[1]} halaman) sedang diproses.")
        return pesanan_dilayani

    def lihat_pesanan_pertama(self):
        return self.antrian[0] if not self.apakah_kosong() else "Antrian kosong"

    def lihat_pesanan_terakhir(self):
        return self.antrian[-1] if not self.apakah_kosong() else "Antrian kosong"

    def apakah_kosong(self):
        return len(self.antrian) == 0

    def jumlah_antrian(self):
        return len(self.antrian)

    def tampilkan_antrian(self):
        if self.apakah_kosong():
            return "Antrian kosong"
        return " <- ".join([f"{id_pesanan} ({jumlah_halaman} hlm)" for id_pesanan, jumlah_halaman in self.antrian])

antrian = AntrianFotokopi()

while True:
    print("SISTEM ANTRIAN LAYANAN FOTOKOPI")
    print("1. Tambah Pesanan Fotokopi")
    print("2. Layani Pesanan (Hapus dari antrian)")
    print("3. Lihat Antrian Saat Ini")
    print("4. Lihat Pesanan di Depan & Belakang")
    print("5. Lihat Jumlah Pesanan dalam Antrian")
    print("6. Keluar")

    pilihan = input("Pilih menu (1-6): ")

    if pilihan == "1":
        id_pesanan = input("Masukkan ID Pesanan (contoh: ZUR57): ").upper()
        jumlah_halaman = int(input("Masukkan jumlah halaman: "))
        antrian.tambah_pesanan(id_pesanan, jumlah_halaman)

    elif pilihan == "2":
        antrian.layani_pesanan()

    elif pilihan == "3":
        print(antrian.tampilkan_antrian())

    elif pilihan == "4":
        print(f"Pesanan di depan: {antrian.lihat_pesanan_pertama()}")
        print(f"Pesanan di belakang: {antrian.lihat_pesanan_terakhir()}")

    elif pilihan == "5":
        print(f"Jumlah pesanan dalam antrian: {antrian.jumlah_antrian()}")

    elif pilihan == "6":
        print("Keluar dari sistem. Terima kasih.")
        break

    else:
        print("Pilihan tidak valid! Silakan masukkan angka 1-6.")
```

```python
class AntrianParkiran:
    def __init__(self, kapasitas=15):
        self.antrian = []
        self.kapasitas = kapasitas

    def tambah_kendaraan(self, nomor_plat, jenis_kendaraan):
        if len(self.antrian) >= self.kapasitas:
            print(f"Parkiran penuh! Kendaraan {nomor_plat} harus menunggu.")
            return

        if "Listrik" in jenis_kendaraan:
            self.antrian.insert(0, (nomor_plat, jenis_kendaraan))
            print(f"Kendaraan {nomor_plat} (Listrik) mendapat prioritas masuk parkiran.")
        else:
            self.antrian.append((nomor_plat, jenis_kendaraan))
            print(f"Kendaraan {nomor_plat} masuk antrian parkir.")

    def keluarkan_kendaraan(self):
        if self.apakah_kosong():
            print("Parkiran kosong! Tidak ada kendaraan untuk keluar.")
            return None
        kendaraan_keluar = self.antrian.pop(0)
        print(f"Kendaraan {kendaraan_keluar[0]} ({kendaraan_keluar[1]}) telah keluar dari parkiran.")
        return kendaraan_keluar

    def lihat_kendaraan_pertama(self):
        return self.antrian[0] if not self.apakah_kosong() else "Parkiran kosong"

    def lihat_kendaraan_terakhir(self):
        return self.antrian[-1] if not self.apakah_kosong() else "Parkiran kosong"

    def apakah_kosong(self):
        return len(self.antrian) == 0

    def jumlah_kendaraan(self):
        return len(self.antrian)

    def tampilkan_antrian(self):
        if self.apakah_kosong():
            return "Parkiran kosong"
        return " <- ".join([f"{nomor_plat} ({jenis})" for nomor_plat, jenis in self.antrian])

parkiran = AntrianParkiran()

while True:
    print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
    print("SISTEM ANTRIAN PARKIRAN MALL")
    print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
    print("1. Tambah Kendaraan ke Parkiran")
    print("2. Keluarkan Kendaraan dari Parkiran")
    print("3. Lihat Antrian Kendaraan")
    print("4. Lihat Kendaraan di Depan & Belakang")
    print("5. Lihat Jumlah Kendaraan dalam Parkiran")
    print("6. Keluar")

    pilihan = input("Pilih menu (1-6): ")

    if pilihan == "1":
        nomor_plat = input("Masukkan Nomor Plat Kendaraan (contoh: KE57): ").upper()
        jenis_kendaraan = input("Masukkan Jenis Kendaraan (Bensin/Diesel/Listrik): ").capitalize()
        parkiran.tambah_kendaraan(nomor_plat, jenis_kendaraan)

    elif pilihan == "2":
        parkiran.keluarkan_kendaraan()

    elif pilihan == "3":
        print(parkiran.tampilkan_antrian())

    elif pilihan == "4":
        print(f"Kendaraan di depan: {parkiran.lihat_kendaraan_pertama()}")
        print(f"Kendaraan di belakang: {parkiran.lihat_kendaraan_terakhir()}")

    elif pilihan == "5":
        print(f"Jumlah kendaraan dalam parkiran: {parkiran.jumlah_kendaraan()}")

    elif pilihan == "6":
        print("Keluar dari sistem. Terima kasih.")
        break

    else:
        print("Pilihan tidak valid! Silakan masukkan angka 1-6.")
```

```python
class AntrianSertifikat:
    def __init__(self, kapasitas=20):
        self.antrian_reguler = []
        self.antrian_prestasi = []
        self.kapasitas = kapasitas

    def tambah_mahasiswa(self, student_id, prestasi):
        total_antrian = len(self.antrian_reguler) + len(self.antrian_prestasi)

        if prestasi.lower() == "ya":
            self.antrian_prestasi.insert(0, student_id)
            print(f"Mahasiswa {student_id} (Prestasi) mendapatkan prioritas pengambilan sertifikat.")
        elif total_antrian >= self.kapasitas:
            print(f"Antrian penuh! Mahasiswa {student_id} harus menunggu.")
        else:
            self.antrian_reguler.append(student_id)
            print(f"Mahasiswa {student_id} masuk ke antrian reguler.")

    def layani_mahasiswa(self):
        if self.apakah_kosong():
            print("Antrian kosong! Tidak ada mahasiswa yang bisa dilayani.")
            return None

        if self.antrian_prestasi:
            mahasiswa_keluar = self.antrian_prestasi.pop(0)
            print(f"Mahasiswa {mahasiswa_keluar} (Prestasi) telah mengambil sertifikat.")
        else:
            mahasiswa_keluar = self.antrian_reguler.pop(0)
            print(f"Mahasiswa {mahasiswa_keluar} (Reguler) telah mengambil sertifikat.")

    def lihat_mahasiswa_pertama(self):
        if self.antrian_prestasi:
            return self.antrian_prestasi[0]
        elif self.antrian_reguler:
            return self.antrian_reguler[0]
        else:
            return "Antrian kosong"

    def lihat_mahasiswa_terakhir(self):
        if self.antrian_reguler:
            return self.antrian_reguler[-1]
        elif self.antrian_prestasi:
            return self.antrian_prestasi[-1]
        else:
            return "Antrian kosong"

    def apakah_kosong(self):
        return len(self.antrian_reguler) + len(self.antrian_prestasi) == 0

    def jumlah_mahasiswa(self):
        return len(self.antrian_reguler) + len(self.antrian_prestasi)

    def tampilkan_antrian(self):
        if self.apakah_kosong():
            return "Antrian kosong"
        return f"Prestasi: {self.antrian_prestasi} | Reguler: {self.antrian_reguler}"


antrian = AntrianSertifikat()

while True:
    print("*************************************")
    print("SISTEM ANTRIAN PENGAMBILAN SERTIFIKAT")
    print("*************************************")
    print("1. Tambah Mahasiswa ke Antrian")
    print("2. Layani Mahasiswa")
    print("3. Lihat Antrian Mahasiswa")
    print("4. Lihat Mahasiswa di Depan & Belakang")
    print("5. Lihat Jumlah Mahasiswa dalam Antrian")
    print("6. Keluar")

    pilihan = input("Pilih menu (1-6): ")

    if pilihan == "1":
        student_id = input("Masukkan ID Mahasiswa (contoh: T57): ").upper()
        prestasi = input("Apakah mahasiswa pernah mengikuti kompetisi tingkat nasional? (Ya/Tidak): ")
        antrian.tambah_mahasiswa(student_id, prestasi)

    elif pilihan == "2":
        antrian.layani_mahasiswa()

    elif pilihan == "3":
        print(antrian.tampilkan_antrian())

    elif pilihan == "4":
        print(f"Mahasiswa di depan: {antrian.lihat_mahasiswa_pertama()}")
        print(f"Mahasiswa di belakang: {antrian.lihat_mahasiswa_terakhir()}")

    elif pilihan == "5":
        print(f"Jumlah mahasiswa dalam antrian: {antrian.jumlah_mahasiswa()}")

    elif pilihan == "6":
        print("Keluar dari sistem. Terima kasih.")
        break

    else:
        print("Pilihan tidak valid! Silakan masukkan angka 1-6.")
```
